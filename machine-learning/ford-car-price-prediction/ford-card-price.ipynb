```python
numeric_columns = ['year', 'price', 'mileage', 'tax', 'mpg', 'engineSize']
```

```python
for column in numeric_columns:
    plt.figure(figsize=(5, 5))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
for column in numeric_columns:
    plt.figure(figsize=(5, 5))
    sns.boxplot(x = df[column])
    plt.title(f'Distribution of {column}')
    plt.show()
```
