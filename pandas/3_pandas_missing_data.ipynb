# Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
df.isnull()
```

# Removing Missing Data

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=1)
```

# Filling Missing Data

```python
df.fillna(value=0)
```

```python
value={ 'A': 10, 'B': 20, 'C': 30, 'D': 40 }
df.fillna(value=value)
```

```python
df.fillna(df.mean())
```
