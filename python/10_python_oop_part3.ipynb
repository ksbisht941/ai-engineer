## 1. Class Relationships

- **Aggregation (Has-A)**: One class uses another class (e.g., A `Project` has many `Tasks`). If the container is destroyed, the contained object can survive.
- **Inheritance (Is-A)**: One class derives properties from another (e.g., a `Dog` is an `Animal`).

```python
# Aggregation Example
class Address:
    def __init__(self, city, state):
        self.city = city
        self.state = state

class Customer:
    def __init__(self, name, address):
        self.name = name
        self.address = address # customer HAS-A address

addr = Address("San Francisco", "CA")
cust = Customer("Kevin", addr)
print(f"{cust.name} lives in {cust.address.city}")
```

## 2. Inheritance and Super keyword

Inheritance allows a child class to inherit attributes and methods from a parent class. The `super()` keyword is used to call methods from the parent class.

```python
class User:
    def __init__(self, username):
        self.username = username

    def login(self):
        print(f"{self.username} logged in")

class Admin(User): # Admin IS-A User
    def __init__(self, username, permissions):
        super().__init__(username) # Calling parent constructor
        self.permissions = permissions

    # Method Overriding
    def login(self):
        super().login() # Call parent login
        print("Admin dashboard unlocked")

adm = Admin("super_kevin", ["all"])
adm.login()
```

## 3. Types of Inheritance

- **Single**: A -> B
- **Multilevel**: A -> B -> C
- **Hierarchical**: A -> B, A -> C
- **Multiple**: A, B -> C (Possible in Python)
- **Hybrid**: Combination of above.

## 4. Multiple Inheritance and Diamond Problem

When a class inherits from multiple classes that share a common ancestor, Python uses **Method Resolution Order (MRO)** to solve the ambiguity.

```python
class A:
    def m(self): print("A")

class B(A):
    def m(self): print("B")

class C(A):
    def m(self): print("C")

class D(B, C):
    pass

obj = D()
obj.m() # Which one? B or C?
print(D.mro()) # Method Resolution Order
```

## 5. Polymorphism

- **Method Overriding**: Child implementation replaces parent (shown above).
- **Method Overloading**: Not strictly supported in Python, but achieved via default arguments or `*args`.

```python
# Pythonic Method Overloading
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

calc = Calculator()
print(calc.add(1, 2))
print(calc.add(1, 2, 3))
```
