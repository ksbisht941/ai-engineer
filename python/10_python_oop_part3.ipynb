## 1. Class Relationships

- **Aggregation (Has-A)**: One class uses another class (e.g., A `Project` has many `Tasks`). If the container is destroyed, the contained object can survive.
- **Inheritance (Is-A)**: One class derives properties from another (e.g., a `Dog` is an `Animal`).

```python
# Aggregation Example
class Address:
    def __init__(self, city, state):
        self.city = city
        self.state = state

class Customer:
    def __init__(self, name, address):
        self.name = name
        self.address = address # customer HAS-A address

addr = Address("San Francisco", "CA")
cust = Customer("Kevin", addr)
print(f"{cust.name} lives in {cust.address.city}")
```

## 2. Inheritance and Super keyword

Inheritance allows a child class to inherit attributes and methods from a parent class. The `super()` keyword is used to call methods from the parent class.

```python
class User:
    def __init__(self, username):
        self.username = username

    def login(self):
        print(f"{self.username} logged in")

class Admin(User): # Admin IS-A User
    def __init__(self, username, permissions):
        super().__init__(username) # Calling parent constructor
        self.permissions = permissions

    # Method Overriding
    def login(self):
        super().login() # Call parent login
        print("Admin dashboard unlocked")

adm = Admin("super_kevin", ["all"])
adm.login()
```

## 3. Types of Inheritance

- **Single**: A -> B
- **Multilevel**: A -> B -> C
- **Hierarchical**: A -> B, A -> C
- **Multiple**: A, B -> C (Possible in Python)
- **Hybrid**: Combination of above.

## 4. Multiple Inheritance and Diamond Problem

When a class inherits from multiple classes that share a common ancestor, Python uses **Method Resolution Order (MRO)** to solve the ambiguity.

```python
class A:
    def m(self): print("A")

class B(A):
    def m(self): print("B")

class C(A):
    def m(self): print("C")

class D(B, C):
    pass

obj = D()
obj.m() # Which one? B or C?
print(D.mro()) # Method Resolution Order
```

## 5. Polymorphism

- **Method Overriding**: Child implementation replaces parent (shown above).
- **Method Overloading**: Not strictly supported in Python, but achieved via default arguments or `*args`.

```python
# Pythonic Method Overloading
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

calc = Calculator()
print(calc.add(1, 2))
print(calc.add(1, 2, 3))
```

## 6. Abstraction

Abstraction allows you to define a "contract" or a blueprint without specifying the details. It helps in managing complexity by letting you work with high-level concepts.

### Abstract Classes and Methods

In Python, we use the `abc` (Abstract Base Classes) module to create abstract classes. 
- **Abstract Class**: A class that contains one or more abstract methods. It cannot be instantiated.
- **Abstract Method**: A method that has a declaration but no implementation in the base class. Derived classes MUST implement these methods.

```python
from abc import ABC, abstractmethod

class BankApp(ABC): # Abstract Class
    def __init__(self):
        print("Bank App initialized")

    @abstractmethod
    def login(self):
        pass # Child must implement login logic

    @abstractmethod
    def transaction(self):
        pass # Child must implement transaction logic

    def info(self):
        print("This is a CineSage Bank Application") # Normal method

# try_create = BankApp() # This would raise TypeError
```

### Implementing the Abstract Class

Now we create concrete classes that implement the abstract methods.

```python
class MobileApp(BankApp):
    def login(self):
        print("Login via Fingerprint/FaceID")
    
    def transaction(self):
        print("Transaction via UPI/QR")

class WebApp(BankApp):
    def login(self):
        print("Login via Username/Password")
    
    def transaction(self):
        print("Transaction via Net Banking")

m = MobileApp()
m.login()

w = WebApp()
w.transaction()
```

### Why use Abstraction?

- **Enforces Structure**: Ensures all child classes follow the same interface.
- **Security**: Hides complex internal logic.
- **Scalability**: Easier to add new types of applications (e.g., DesktopApp) without breaking existing code.
