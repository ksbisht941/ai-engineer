Object-Oriented Programming (OOP) is a programming paradigm that relies on the concept of classes and objects.

## 1. What is OOP?

OOP is a style of programming that allows you to group data (attributes) and behaviors (methods) into structures called **Classes**.
- **Class**: A blueprint for creating objects (e.g., a "Bank Account" template).
- **Object**: An instance of a class (e.g., "Kevin's actual account").

## 2. Banking Application Code

Demonstrating methods, functions, and the `self` keyword.

```python
class BankAccount:
    # Constructor - runs automatically when object is created
    def __init__(self, owner, balance=0):
        self.owner = owner     # Attribute
        self.balance = balance # Attribute

    # Method - function belonging to a class
    def deposit(self, amount):
        self.balance += amount
        return f"Deposited {amount}. New balance: {self.balance}"

    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient funds"
        self.balance -= amount
        return f"Withdrew {amount}. New balance: {self.balance}"

# Creating an object
acc = BankAccount("Kevin", 1000)
print(acc.deposit(500))
print(acc.withdraw(200))
```

## 3. The Concept of 'self'

`self` represents the instance of the class. It allows you to access variables that belong to the class. It is always the first parameter of any class method.

## 4. Magic / Dunder Methods

Dunder (Double underscore) methods allow you to implement operator overloading and special behaviors (like how an object is printed).

## 5. Fraction Class Example

Implementing custom logic for printing and arithmetic.

```python
class Fraction:
    def __init__(self, n, d):
        self.num = n
        self.den = d

    # __str__: How the object is printed
    def __str__(self):
        return f"{self.num}/{self.den}"

    # __add__: Overloading the + operator
    def __add__(self, other):
        temp_num = self.num * other.den + other.num * self.den
        temp_den = self.den * other.den
        return f"{temp_num}/{temp_den}"

    # __sub__: Overloading the - operator
    def __sub__(self, other):
        temp_num = self.num * other.den - other.num * self.den
        temp_den = self.den * other.den
        return f"{temp_num}/{temp_den}"

    # __mul__: Overloading the * operator
    def __mul__(self, other):
        temp_num = self.num * other.num
        temp_den = self.den * other.den
        return f"{temp_num}/{temp_den}"

    # __truediv__: Overloading the / operator
    def __truediv__(self, other):
        temp_num = self.num * other.den
        temp_den = self.den * other.num
        return f"{temp_num}/{temp_den}"

f1 = Fraction(1, 2)
f2 = Fraction(1, 3)

print(f"f1: {f1}, f2: {f2}")
print(f"Addition: {f1 + f2}")
print(f"Subtraction: {f1 - f2}")
print(f"Multiplication: {f1 * f2}")
print(f"Division: {f1 / f2}")
```
