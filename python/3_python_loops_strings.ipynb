# Session 3: Python Strings + Advanced Loops

This notebook covers loop control statements (`break`, `continue`, `pass`) and an in-depth look at Python Strings.

## 1. Loop Control Statements

These statements change the execution flow of a loop.

```python
# break: exits the loop
for i in range(1, 10):
    if i == 5:
        break
    print(i)

print("---")

# continue: skips the current iteration
for i in range(1, 10):
    if i % 2 == 0:
        continue
    print(i)

print("---")

# pass: null statement (does nothing)
for i in range(1, 5):
    pass # will be implemented later
```

## 2. Solving Loop Problems

Example: Finding if a number is Prime.

```python
number = 7
is_prime = True

for i in range(2, number):
    if number % i == 0:
        is_prime = False
        break

if is_prime:
    print(f"{number} is a prime number")
else:
    print(f"{number} is not a prime number")
```

## 3. Python Strings

Strings are sequences of characters wrapped in quotes.

```python
s = "Hello World"
print(s)
```

## 4. String Indexing and Slicing

Accessing parts of a string.

```python
s = "Python"

# Indexing
print(s[0])  # First character
print(s[-1]) # Last character

# Slicing [start:stop:step]
print(s[0:3])   # First three characters
print(s[2:])    # From index 2 to end
print(s[::-1])  # Reverse string
```

## 5. String Immutability

Strings cannot be changed once created. You must create a new string.

```python
s = "Hello"
# s[0] = 'M' # This would cause an error

# To "edit", create a new one
s = "M" + s[1:]
print(s)

# Delete a string reference
del s
```

## 6. Operations on Strings

Common arithmetic and membership operations.

```python
s1 = "Hello"
s2 = "World"

# Concatenation
print(s1 + " " + s2)

# Repetition
print("Hi" * 3)

# Membership
print("H" in "Hello")
print("z" not in "Hello")
```
