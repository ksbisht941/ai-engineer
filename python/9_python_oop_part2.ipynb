## 1. Attribute Access and Dynamic Creation

In Python, attributes can be accessed via `self` inside the class or via the object outside the class. You can even create new attributes dynamically from outside.

```python
class Person:
    def __init__(self, name):
        self.name = name

p1 = Person("Kevin")
print(p1.name) # Accessing from outside

# Dynamic Attribute Creation
p1.age = 25
print(p1.age)
```

## 2. Reference Variables

When you assign an object to a variable, you are storing a **reference** to that object in memory, not the object itself.

```python
p2 = p1 # p2 now points to the same object as p1
p2.name = "Bob"
print(p1.name) # Also changes p1 because they are references to the same data
```

## 3. Encapsulation

Encapsulation is the practice of hiding internal details. In Python, we use a double underscore `__` to make attributes "private".

```python
class Student:
    def __init__(self, name, pin):
        self.name = name
        self.__pin = pin # Private attribute

    def get_pin(self):
        return self.__pin # Getter method
    
    def set_pin(self, new_pin):
        self.__pin = new_pin # Setter method

s = Student("Alice", 1234)
print(s.name)
# print(s.__pin) # This would raise an AttributeError
print(s.get_pin())
```

## 4. Static Variables and Methods

- **Static Variables**: Shared across all instances of a class.
- **Static Methods**: Functions that don't need access to instance attributes (`self`).

```python
class ATM:
    counter = 1 # Static/Class Variable

    def __init__(self):
        self.serial = ATM.counter
        ATM.counter += 1

    @staticmethod
    def get_bank_info():
        return "CineSage Bank - 24/7 Service"

t1 = ATM()
t2 = ATM()
print(t2.serial) # Outputs 2
print(ATM.get_bank_info())
```
