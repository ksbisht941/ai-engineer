# Session 2: Python Operators + if-else + Loops

This notebook covers operators, conditional logic, modules, and loops in Python.

## 1. Python Operators

Operators are used to perform operations on variables and values.

```python
# Arithmetic Operators
print(10 + 5) # Addition
print(10 - 5) # Subtraction
print(10 * 5) # Multiplication
print(10 / 5) # Division
print(10 % 3) # Modulus (Remainder)
print(10 ** 2) # Exponent (Power)
print(10 // 3) # Floor Division (Quotient)

# Comparison Operators
print(10 == 10) # Equal
print(10 != 5)  # Not Equal
print(10 > 5)   # Greater than
print(10 < 20)  # Less than

# Logical Operators
print(True and False) # Logical AND
print(True or False)  # Logical OR
print(not True)       # Logical NOT
```

## 2. Python If-Else

Conditional statements allow you to execute code based on conditions.

```python
age = 20

if age >= 18:
    print("You are an adult") # runs if condition is true
elif age > 12:
    print("You are a teenager") # runs if first condition is false
else:
    print("You are a child") # runs if all conditions are false
```

## 3. Python Modules

Modules are files containing Python code that can be imported.

```python
import math # Importing the math module

print(math.sqrt(16)) # using a function from the module
print(math.pi)       # using a constant from the module
```

## 4. Python While Loop

A `while` loop repeats a block of code as long as a condition is true.

```python
count = 1
while count <= 5:
    print("Count is:", count)
    count += 1 # incrementing the counter
```

## 5. Python For Loop

A `for` loop is used to iterate over a sequence (list, string, range, etc.).

```python
# Iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Using range()
for i in range(3):
    print("Index:", i)
```
