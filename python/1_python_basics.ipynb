## 1. About Python

Python is a high-level, interpreted, and general-purpose programming language. It is known for its simplicity and readability.

## 2. Python Output / print() function

The `print()` function is used to display output on the screen.

```python
# Simple print statement
print("Hello World")

# Printing multiple items
print("Python", "is", "fun")
```

## 3. Python Data Types

Python has several built-in data types:

```python
# Integer
print(type(5))

# Float
print(type(5.5))

# String
print(type("Hello"))

# Boolean
print(type(True))

# List
print(type([1, 2, 3]))

# Dictionary
print(type({"name": "Kevin"}))
```

## 4. Python Variables

Variables are containers for storing data values.

```python
# Creating a variable
name = "Kevin"
age = 25

print(name)
print(age)
```

## 5. Python Comments

Comments are used to explain the code and make it more readable.

```python
# This is a single line comment
print("Comments are ignored by Python") # This is also a comment
```

## 6. Python Keywords and Identifiers

- **Keywords**: Reserved words that cannot be used as variable names (e.g., `if`, `else`, `print`).
- **Identifiers**: Names given to entities like variables, functions, etc. (e.g., `my_var`).

## 7. Python User Input

The `input()` function allows user input.

```python
# Taking input from user
# comment: input() always returns a string
name = "Kevin" # Simulation of input
print("Hello " + name)
```

## 8. Python Type Conversion

Converting one data type to another.

```python
# Convert string to integer
num_str = "10"
num_int = int(num_str)
print(type(num_int))

# Convert integer to float
print(float(5))
```

## 9. Python Literals

Literals are raw data assigned to variables (e.g., `10`, `"Hello"`, `3.14`).

```python
a = 0b1010 # Binary Literal
b = 100 # Decimal Literal
c = 0o310 # Octal Literal
d = 0x12c # Hexadecimal Literal

print(a, b, c, d)
```
