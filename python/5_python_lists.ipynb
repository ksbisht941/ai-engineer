# Session 4: Python Lists

This notebook covers everything about Python Lists - from memory management to advanced operations.

## 1. Introduction to Lists

- **Arrays vs Lists**: Arrays are homogeneous (store same types), Lists are heterogeneous (can store different types).
- **Memory**: Lists in Python are referential. They store addresses of objects rather than raw data.
- **Characteristics**: Mutable, Ordered, Heterogeneous, Dynamic Scaling.

## 2. Create and Access

Creating lists and using indexing.

```python
L = [1, 2, 3, "Hello", True] # Heterogeneous list
print(L[0])  # Access first item
print(L[-1]) # Access last item
```

## 3. Adding Items: append, extend, insert

- `append(val)`: Adds one item at the end.
- `extend([vals])`: Adds multiple items at the end.
- `insert(index, val)`: Adds one item at a specific index.

```python
L = [1, 2]
L.append(3)
L.extend([4, 5])
L.insert(1, "Middle")
print(L)
```

## 4. Editing and Deleting items

Lists are mutable, so we can change items.

```python
L = [1, 2, 3]
L[0] = 100 # Edit

del L[1]      # Delete by index
L.remove(100) # Delete by value
item = L.pop() # Delete and return last item
print("Remaining:", L)
```

## 5. Operations: Arithmetic, Membership, Loops

Using operators with lists.

```python
L1 = [1, 2]
L2 = [3, 4]

print(L1 + L2) # Concatenation
print(L1 * 3)  # Repetition
print(1 in L1) # Membership test

for i in L1:
    print("Looping:", i)
```

## 6. List Comprehension

Concise way to create lists.

```python
# Create a list of squares
squares = [x**2 for x in range(1, 6)]
print(squares)
```

## 7. Zip() function

Combining two lists into pairs.

```python
names = ["Kevin", "Bob"]
scores = [90, 85]

for name, score in zip(names, scores):
    print(f"{name} scored {score}")
```

## 8. Disadvantages

- Slower for large data (referential memory access overhead).
- Memory heavy compared to arrays.
