# Session 11: Exception Handling

Errors are inevitable in programming. This notebook covers how to distinguish between syntax errors and exceptions, and how to handle them gracefully.

## 1. Syntax Error vs Exceptions

- **Syntax Error**: Occurs when you write code that violates the rules of the language (e.g., missing a colon). The program won't even start.
- **Exception**: Occurs during the execution of a program (e.g., division by zero). The code is syntactically correct, but something went wrong at runtime.

```python
# Example of Exception (Runtime Error)
# print(1/0) # This would raise ZeroDivisionError
```

## 2. Why handle Exceptions?

Without handling, any error will terminate the program immediately. Handling allows the program to continue or shut down gracefully, providing helpful feedback.

## 3. Try-Except-Else-Finally

- `try`: Code that might raise an error.
- `except`: Code to execute if an error occurs.
- `else`: Code to execute if NO error occurs.
- `finally`: Code that ALWAYS executes (cleanup).

```python
try:
    num = int(input("Enter a number: "))
    result = 100 / num
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
except ValueError:
    print("Error: Please enter a valid integer.")
else:
    print(f"Success! Result is {result}")
finally:
    print("Operation complete.")
```

## 4. Raising Exceptions

You can manually trigger an exception using the `raise` keyword.

```python
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative!")
    return f"Age set to {age}"

try:
    print(check_age(-5))
except ValueError as e:
    print(f"Caught: {e}")
```

## 5. Custom Exceptions

You can create your own error types by inheriting from the `Exception` class.

```python
class BalanceInsufficientError(Exception):
    """Raised when withdrawal exceeds balance."""
    pass

def withdraw(amount, balance):
    if amount > balance:
        raise BalanceInsufficientError(f"Withdrawal of {amount} failed. Balance: {balance}")
    return balance - amount

try:
    withdraw(500, 200)
except BalanceInsufficientError as e:
    print(f"Custom Error: {e}")
```
