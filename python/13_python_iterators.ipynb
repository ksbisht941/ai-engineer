# Session 13: Iterators

Iterators are objects that allow you to traverse through elements in a collection. This notebook explains the difference between iterables and iterators and how to build your own.

## 1. Iterable vs Iterator

- **Iterable**: Any object that can return its members one at a time (e.g., list, string, dict). It implements `__iter__`.
- **Iterator**: An object that represents a stream of data. It implements `__next__` and returns the next element.

```python
nums = [1, 2, 3] # nums is an Iterable
iter_nums = iter(nums) # iter_nums is an Iterator

print(next(iter_nums))
print(next(iter_nums))
print(next(iter_nums))
# print(next(iter_nums)) # This would raise StopIteration
```

## 2. Making Your Own For Loop

A `for` loop in Python works by converting an iterable to an iterator and calling `next()` repeatedly.

```python
def my_for_loop(iterable):
    iterator = iter(iterable)
    while True:
        try:
            item = next(iterator)
            print(f"Looping item: {item}")
        except StopIteration:
            break

my_for_loop([10, 20, 30])
```

## 3. Custom Range Function

Building a custom class that acts as a range iterator.

```python
class MyRange:
    def __init__(self, start, end):
        self.current = start
        self.end = end
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current >= self.end:
            raise StopIteration
        
        value = self.current
        self.current += 1
        return value

for i in MyRange(1, 5):
    print("Custom range value:", i)
```
