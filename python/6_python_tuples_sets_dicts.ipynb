# Session 5: Tuples, Sets, and Dictionaries

This notebook covers the more complex built-in data structures in Python: Tuples, Sets, and Dictionaries.

## 1. Tuples

- **Immutability**: Once created, you cannot change, add, or remove items.
- **Ordered**: Follows a zero-indexed order.
- **Heterogeneous**: Can store different data types.

```python
# Creation and Access
t = (1, 2, 3, "Hello")
print(t[0])

# Immutability (uncomment to see error)
# t[0] = 100 

# Unpacking
a, b, c, d = t
print(a, d)

# Functions
print(len(t))
print(t.index("Hello"))
```

## 2. Sets

- **Unordered**: No indexing or slicing.
- **Unique items**: No duplicates allowed.
- **Mutable**: You can add/remove items (but items themselves must be immutable).

```python
# Creation
s = {1, 2, 3, 3, 3} # Duplicates disappear
print(s)

# Adding and Deleting
s.add(4)
s.remove(1)
print(s)

# Operations (Union, Intersection)
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1 | s2) # Union
print(s1 & s2) # Intersection

# Frozen Set (Immutable set)
fs = frozenset([1, 2, 3])
```

## 3. Dictionaries

- **Key-Value Pairs**: Data is stored as `key: value` relationships.
- **Mutable**: Keys must be unique and immutable; values can be anything.

```python
# Creation
d = {"name": "Kevin", "age": 25}

# Access and Edit
print(d["name"])
d["age"] = 26
print(d)

# Add and Delete
d["city"] = "NYC"
del d["age"]
print(d)

# Dictionary Methods
print(d.keys())
print(d.values())
print(d.items())
```

## 4. Comprehensions and Zip

Advanced ways to create and iterate over these structures.

```python
# Set Comprehension
even_set = {x for x in range(10) if x % 2 == 0}
print(even_set)

# Dictionary Comprehension
squares_dict = {x: x**2 for x in range(1, 5)}
print(squares_dict)

# Zip on Dictionary
keys = ["A", "B"]
vals = [1, 2]
zipped_dict = dict(zip(keys, vals))
print(zipped_dict)
```
