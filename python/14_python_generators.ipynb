# Session 14: Generators

Generators are a simple way to create iterators using functions. They help in processing large datasets by generating values on the fly instead of storing them in memory.

## 1. What is a Generator?

A generator is a function that uses the `yield` keyword instead of `return`. It maintains its state between executions.

```python
def simple_generator():
    yield "First item"
    yield "Second item"
    yield "Third item"

gen = simple_generator()
for val in gen:
    print(val)
```

## 2. Yield vs Return

- `return`: Ends the function and gives back a final value.
- `yield`: Pauses the function, gives back a value, and can resume from the same point later.

## 3. Generator Expressions

Similar to list comprehensions, but uses parenthese `()` to create an iterator rather than a list.

```python
square_gen = (x*x for x in range(1, 10)) # Memory efficient
print(next(square_gen))
print(next(square_gen))
```

## 4. Benefits of Generators

- **Memory Efficiency**: Doesn't store the entire sequence in RAM.
- **Infinite Streams**: Can generate elements forever without consuming infinite memory.
- **Performance**: Lazy evaluation means we only process values as they are needed.

```python
# Infinite Fibonacci sequence generator
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()
for i in range(5):
    print(f"Fibonacci {i+1}: {next(fib)}")
```
