## 1. Function Basics

- **Creation**: Using the `def` keyword.
- **Arguments vs Parameters**: Parameters are variables in the definition; arguments are values passed during the call.

```python
def greet(name, message="Hello"): # message is a default argument
    """This is a docstring describing the function."""
    return f"{message}, {name}!"

print(greet("Kevin"))
print(greet("Bob", "Hi"))

# Accessing documentation
print(greet.__doc__)
```

## 2. *args and **kwargs

- `*args`: Pass a variable number of non-keyword arguments (as a tuple).
- `**kwargs`: Pass a variable number of keyword arguments (as a dictionary).

```python
def sum_all(*args):
    return sum(args)

def show_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print(sum_all(1, 2, 3, 4))
show_info(name="Kevin", age=25, city="San Francisco")
```

## 3. Variable Scope and Nested Functions

- **Local vs Global**: Variables defined inside a function are local.
- **Nested Functions**: Functions defined within functions. They can access outer scope variables (Closures).

```python
global_var = "I am global"

def outer_function(x):
    local_var = "I am local"
    
    def inner_function():
        # Accessing outer variable (Closure)
        return f"{local_var} - processed with {x}"
    
    return inner_function()

print(outer_function("data"))
```

## 4. First-Class Citizens

Python functions can be assigned to variables, passed as arguments, and returned from other functions.

```python
def square(x):
    return x * x

f = square # Assigned to variable
print(f(5))

def apply_func(func, val):
    return func(val) # Passed as argument

print(apply_func(square, 10))
```

## 5. Functional Programming: Lambda, Map, Filter, Reduce

- **Lambda**: Anonymous one-line functions.
- **Map**: Apply function to all items in an iterable.
- **Filter**: Extract items that satisfy a condition.
- **Reduce**: Perform rolling computation on sequential pairs.

```python
from functools import reduce

nums = [1, 2, 3, 4, 5, 6]

# Lambda
double = lambda x: x * 2

# Map
doubled_nums = list(map(lambda x: x * 2, nums))
print("Map:", doubled_nums)

# Filter
even_nums = list(filter(lambda x: x % 2 == 0, nums))
print("Filter:", even_nums)

# Reduce
total_sum = reduce(lambda x, y: x + y, nums)
print("Reduce (Sum):", total_sum)
```
