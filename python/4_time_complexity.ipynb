# Session: Time Complexity

This notebook covers the efficiency of programs and the orders of growth (Big O Notation).

## 1. Efficiency in Programming

Efficiency is about how well a program uses resources (Time and Space). 
- **Time Complexity**: How the execution time increases as the input size grows.
- **Space Complexity**: How the memory usage increases as the input size grows.

## 2. Orders of Growth (Big O)

Common Time Complexities:
- $O(1)$: Constant Time (Accessing a list index)
- $O(\log n)$: Logarithmic Time (Binary Search)
- $O(n)$: Linear Time (Linear Search)
- $O(n^2)$: Quadratic Time (Nested loops)
- $O(2^n)$: Exponential Time (Recursive Fibonacci)

## 3. Examples of Time Complexity

```python
# O(1) - Constant Time
def get_first_element(lst):
    return lst[0] # comment: operation takes same time regardless of list size

print(get_first_element([1, 2, 3]))
```

```python
# O(n) - Linear Time
def print_all_elements(lst):
    for item in lst: # comment: loop runs 'n' times for 'n' items
        print(item)

print_all_elements([10, 20, 30])
```

```python
# O(n^2) - Quadratic Time
def print_pairs(lst):
    for i in lst:           # runs n times
        for j in lst:       # runs n times for each i
            print(i, j)     # total operations: n * n

print_pairs([1, 2])
```
