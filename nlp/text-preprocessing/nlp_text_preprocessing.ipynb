```python
import re
```

```python
customer_review = r"Kevi delivered excellent results despite the impossible deadline pressure yesterday."
customer_review = customer_review.lower()
customer_review
```

```python
new_customer_review = re.sub("kevi", "kevin", customer_review)
new_customer_review
```

```python
customer_reviews = [
    "Kevin delivered excellent results despite the impossible deadline pressure yesterday.",
    "The new feature looked promising but failed miserably in production.",
    "I think Kevin underestimated the risks and caused delays today.",
    "Your presentation was clear, confident, and surprisingly very persuasive overall.",
    "The team ignored critical feedback and made avoidable mistakes again.",
    "Kevin quickly fixed the bug and saved the release schedule.",
    "Honestly, this plan could succeed or completely embarrass us publicly."
]
customer_reviews = [review.lower() for review in customer_reviews]
customer_reviews
```

```python
kevin_reviews = []
```

```python
pattern_to_find = r"kevin?"
```

```python
for review in customer_reviews:
    if re.search(pattern_to_find, review):
        kevin_reviews.append(review)

kevin_reviews
```

```python
start_with_kevin = []
find_start_pattern = r"^kevin"
```

```python
for review in customer_reviews:
    if re.search(find_start_pattern, review):
        start_with_kevin.append(review)

start_with_kevin
```

```python
end_with_yesterday = []
find_end_pattern = r"yesterday.$"
```

```python
for review in customer_reviews:
    if re.search(find_end_pattern, review):
        end_with_yesterday.append(review)

end_with_yesterday
```

```python
text = "Hello World"
match_result = re.match("Hello", text)
print(f"re.match on 'Hello' in '{text}': {match_result.group() if match_result else 'No match'}")

match_result_fail = re.match("World", text)
print(f"re.match on 'World' in '{text}': {match_result_fail.group() if match_result_fail else 'No match'}")
```

```python
text = "Python"
fullmatch_result = re.fullmatch("Python", text)
print(f"re.fullmatch on 'Python' in '{text}': {fullmatch_result.group() if fullmatch_result else 'No match'}")

fullmatch_result_fail = re.fullmatch("Py", text)
print(f"re.fullmatch on 'Py' in '{text}': {fullmatch_result_fail.group() if fullmatch_result_fail else 'No match'}")

fullmatch_result_fail_2 = re.fullmatch("Python Programming", "Python")
print(f"re.fullmatch on 'Python Programming' in 'Python': {fullmatch_result_fail_2.group() if fullmatch_result_fail_2 else 'No match'}")
```

# Practical Task: Text Preprocessing

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
data['review_lowercase'] = data['Review'].str.lower()
data['review_lowercase'][0]
```

```python
en_stopwords.remove('not')
```

```python
data['remove_stopwords'] = data['review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in en_stopwords]))
data['remove_stopwords'][0]
```

```python
data['no_punct'] = data.apply(lambda x: re.sub(r'[*]', 'star', x['remove_stopwords']), axis=1)
data.head()
```

```python
data['no_punct'] = data.apply(lambda x: re.sub(r'([^\w\s])', '', x['no_punct']), axis=1)
data.head()
```

```python
data['tokenized'] = data.apply(lambda x: word_tokenize(x['no_punct']), axis=1)
data.head()
```

```python
data['stemmed'] = data['tokenized'].apply(lambda tokens: [ps.stem(token) for token in tokens])
data.head()
```

```python
data['lemmatized'] = data['tokenized'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
data.head()
```

```python
clean_tokens = sum(data['lemmatized'], [])
clean_tokens
```

```python
unigrams = pd.Series(nltk.ngrams(clean_tokens, 1), name='unigrams').value_counts()
unigrams[:10]
```

```python
bigrams = pd.Series(nltk.ngrams(clean_tokens, 2), name='unigrams').value_counts()
bigrams[:10]
```

```python
trigrams = pd.Series(nltk.ngrams(clean_tokens, 3), name='unigrams').value_counts()
trigrams[:10]
```
