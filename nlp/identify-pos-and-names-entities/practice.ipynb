```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
# import spacy
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
bbc_data = pd.read_csv("bbc_news.csv")
```

```python
bbc_data.head()
```

```python
bbc_data.info()
```

```python
titles = pd.DataFrame(bbc_data['title'])
titles.head()
```

```python
titles['lowercase'] = titles['title'].str.lower()
titles.head()
```

```python
en_stopwords = stopwords.words('english')
titles['no_stopwords'] = titles['lowercase'].apply(lambda x: ' '.join([word for word in x.split(' ') if word not in (en_stopwords)]))
titles.head()
```

```python
titles['no_stopwords_no_punc'] = titles.apply(lambda x: re.sub(r"([^\w\s])", "", x["no_stopwords"]), axis=1)
titles.head()
```

```python
titles['tokens_raw'] = titles.apply(lambda x: word_tokenize(x['title']), axis=1)
titles['tokens_clean'] = titles.apply(lambda x: word_tokenize(x['no_stopwords_no_punc']), axis=1)
titles.head()
```

```python
lemmatizer = WordNetLemmatizer()
titles['tokens_clean_lemmatized'] = titles['tokens_clean'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
titles.head()
```

```python
tokens_raw_list = sum(titles['tokens_raw'], [])
tokens_raw_list
```

```python
tokens_clean_list = sum(titles['tokens_clean'], [])
tokens_clean_list
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
spacy_doc = nlp(''.join(tokens_raw_list))
```

```python
pos_df = pd.DataFrame(columns=['token', 'pos_tag'])

for token in spacy_doc:
    # print(token)
    pos_df = pd.concat([pos_df,
                        pd.DataFrame.from_records([{'token': token, 'pos_tag': token.pos_}])], ignore_index=True)
```
