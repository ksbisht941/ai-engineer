```python
from ultralytics import YOLO
```

```python
# Load a model
model = YOLO("yolo26n.yaml")  # build a new model from YAML
model = YOLO("yolo26n.pt")  # load a pretrained model (recommended for training)
model = YOLO("yolo26n.yaml").load("yolo26n.pt")  # build from YAML and transfer weights
```

```python
def detect(image_path, conf_threshold=0.5):
    results = model(image_path)

    output = []
    for r in results:
        for box in r.boxes:
            if float(box.conf) >= conf_threshold:
                output.append({
                    "label": model.names[int(box.cls)],
                    "confidence": float(box.conf),
                    "bbox": box.xyxy.tolist()
                })

    return output
```

```python
detect("dataset/bus.jpg")
```

```python
def train():

    model.train(data="coco8.yaml", epochs=100, imgsz=640)

if __name__ == "__main__":
    train()
```

```python
class YOLODetector:
    def __init__(self, model_path):
        self.model = YOLO(model_path)

    def predict(self, image_path, conf=0.5):
        results = self.model(image_path)

        detections = []
        for r in results:
            for box in r.boxes:
                if float(box.conf) >= conf:
                    detections.append({
                        "label": self.model.names[int(box.cls)],
                        "confidence": float(box.conf),
                        "bbox": box.xyxy.tolist()
                    })

        return detections
```

```python
detector = YOLODetector("runs/detect/custom_yolo/weights/best.pt")

results = detector.predict("test.jpg")

print(results)
```
