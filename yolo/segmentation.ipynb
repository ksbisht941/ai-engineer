```python
from ultralytics import YOLO
```

```python
# Load a model
model = YOLO("yolo26n-seg.yaml")  # build a new model from YAML
model = YOLO("yolo26n-seg.pt")  # load a pretrained model (recommended for training)
model = YOLO("yolo26n-seg.yaml").load("yolo26n.pt")  # build from YAML and transfer weights
```

```python
def train_segmentation():
    # Train the model
    results = model.train(data="coco8-seg.yaml", epochs=100, imgsz=640)

if __name__ == "__main__":
    train_segmentation()
```

```python
class YOLOSegmentation:
    def __init__(self, model_path):
        self.model = YOLO(model_path)

    def predict(self, image_path, conf=0.5):
        results = self.model(image_path)

        output = []
        for r in results:
            for i, box in enumerate(r.boxes):
                if float(box.conf) >= conf:
                    mask = r.masks.data[i].cpu().numpy()

                    output.append({
                        "label": self.model.names[int(box.cls)],
                        "confidence": float(box.conf),
                        "bbox": box.xyxy.tolist(),
                        "mask": mask.tolist()  # pixel mask
                    })

        return output
```

```python
seg = YOLOSegmentation("runs/segment/seg_model/weights/best.pt")

results = seg.predict("dataset/bus.jpg")
```
