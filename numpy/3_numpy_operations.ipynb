# Advance Array Manipulation

```python
arr1 = np.arange(1, 6)
arr2 = np.arange(6, 11)
```

```python
np.vstack((arr1, arr2))
```

```python
np.hstack((arr1, arr2))
```

```python
np.column_stack((arr1, arr2))
```

```python
arr1 = np.arange(1, 37).reshape(6, 6)
arr1
```
