```python
import numpy as np
```

# Reshape and Resizing

```python
arr = np.arange(1, 31)
arr
```

```python
arr.reshape(5, 6)
```

```python
arr.resize(5, 6, refcheck=False)
arr
```

### Key Differences Summarized:

*   **`reshape()`**: Returns a *new* array; the original array is unchanged. The number of elements must be the same.
*   **`resize()`**: Modifies the array *in-place*. The number of elements can change (truncated or padded with zeros). Be careful with `refcheck` if other variables reference the array.

# Indexing and Slicing of Vectors

```python
arr = np.arange(11, 21)
arr
```

```python
arr[3::2]
```

# Indexing and Slicing of Matrix

```python
arr = np.arange(1, 31).reshape(6, 5)
arr
```

```python
arr[3:, 3:]
```

```python
arr[:, 2]
```

# Boolean Indexing

```python
arr = np.arange(11, 21)
arr
```

```python
bool_idx = arr % 2 == 0
bool_idx
```

```python
arr = arr[bool_idx]
arr
```
