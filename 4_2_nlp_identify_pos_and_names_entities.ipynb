```python
import spacy
import pandas as pd
```

```python
nlp = spacy.load("en_core_web_sm")
nlp
```

```python
paragraph = "Books are an important part of our life. We read many kinds of books. We study text books at school that are part of our syllabus. \nWe also have story books that we read for leisure and fun. We can get them from the school library. We can also buy them from bookshops. Books are printed on paper, and can have pictures too that make them interesting. We can read books on the computer as well. \nI love reading children’s stories published in colourful books, as they are beautifully presented. I have a nice collection of story books at home."
paragraph
```

```python
spacy_doc = nlp(paragraph)
spacy_doc
```

```python
pos_df = pd.DataFrame(columns=['Token', 'POS'])
```

```python
for token in spacy_doc:
    # print(token)
    pos_df = pd.concat([pos_df,
                        pd.DataFrame.from_records([{'Token': token, 'POS': token.pos_}])], ignore_index=True)
```
