# OpenCV Drawing and Text

This notebook covers drawing shapes (lines, rectangles, circles) and adding text overlays.

```python
import cv2
image = cv2.imread("input_img/sample.jpg")
```

## 1. Drawing Lines

```python
# Draw a green line on the image
if image is not None:
    pt1 = (50, 300)
    pt2 = (300, 300)
    color = (0, 255, 0) # Green in BGR
    thickness = 4
    
    line_img = cv2.line(image.copy(), pt1, pt2, color, thickness)
    cv2.imwrite("output_img/line.jpg", line_img)
else:
    print("Image not found.")
```

## 2. Drawing Rectangles

```python
# Draw a green rectangle on the image
if image is not None:
    pt1 = (50, 50)
    pt2 = (250, 250)
    color = (0, 255, 0)
    thickness = 5
    
    rect_img = cv2.rectangle(image.copy(), pt1, pt2, color, thickness)
    cv2.imwrite("output_img/rectangle.jpg", rect_img)
else:
    print("Image not found.")
```

## 3. Drawing Circles

```python
# Draw a green circle on the image
if image is not None:
    center = (256, 256)
    radius = 100
    color = (0, 255, 0)
    thickness = -1 # Filled circle
    
    circle_img = cv2.circle(image.copy(), center, radius, color, thickness)
    cv2.imwrite("output_img/circle.jpg", circle_img)
else:
    print("Image not found.")
```

## 4. Adding Text

```python
# Add text to the image
if image is not None:
    text = "Hello World!"
    pos = (30, 250)
    font = cv2.FONT_HERSHEY_SIMPLEX
    scale = 1.5
    color = (255, 255, 255)
    thickness = 2
    
    text_img = cv2.putText(image.copy(), text, pos, font, scale, color, thickness, cv2.LINE_AA)
    cv2.imwrite("output_img/text.jpg", text_img)
else:
    print("Image not found.")
```
