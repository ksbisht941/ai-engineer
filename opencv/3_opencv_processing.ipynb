# OpenCV Image Processing

This notebook covers blurring, edge detection, and thresholding.

```python
import cv2
import numpy as np

image = cv2.imread("input_img/sample.jpg")
```

## 1. Blurring

```python
# Apply Gaussian blur with a (15, 15) kernel
if image is not None:
    blurred_img = cv2.GaussianBlur(image.copy(), (15, 15), 0)
    cv2.imwrite("output_img/blurred.jpg", blurred_img)
else:
    print("Image not found.")
```

## 2. Edge Detection

```python
# Detect edges using the Canny algorithm
if image is not None:
    edges = cv2.Canny(image.copy(), 100, 300)
    cv2.imwrite("output_img/canny.jpg", edges)
else:
    print("Image not found.")
```

## 3. Thresholding

```python
# Apply simple binary thresholding to a grayscale image
if image is not None:
    greyscale_image = cv2.cvtColor(image.copy(), cv2.COLOR_BGR2GRAY)
    
    thresh_value = 127
    max_value = 255
    ret, thresh = cv2.threshold(greyscale_image, thresh_value, max_value, cv2.THRESH_BINARY)
    cv2.imwrite("output_img/threshold.jpg", thresh)
else:
    print("Image not found.")
```

# 4. Dilate

```python
if image is not None:
    # Load the image as grayscale
    greyscale_image = cv2.cvtColor(image.copy(), cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding to ensure foreground is white (optional but recommended)
    ret, thresh = cv2.threshold(greyscale_image, 127, 255, cv2.THRESH_BINARY_INV) # Inverse threshold may be needed

    # Define a 5x5 kernel
    kernel = np.ones((5,5), np.uint8)

    # Apply dilation with 2 iterations
    dilated_image = cv2.dilate(thresh, kernel, iterations = 2)

    cv2.imwrite("output_img/dilate.jpg", dilated_image)
else:
    print("Image not found.")
```
