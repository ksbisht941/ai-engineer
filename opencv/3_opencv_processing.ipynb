# OpenCV Image Processing

This notebook covers blurring, edge detection, and thresholding.

```python
import cv2
image = cv2.imread("input_img/sample.jpg")
```

## 1. Blurring

```python
# Apply Gaussian blur with a (15, 15) kernel
if image is not None:
    blurred_img = cv2.GaussianBlur(image, (15, 15), 0)
    cv2.imwrite("output_img/blurred.jpg", blurred_img)
else:
    print("Image not found.")
```

## 2. Edge Detection

```python
# Detect edges using the Canny algorithm
if image is not None:
    edges = cv2.Canny(image, 100, 300)
    cv2.imwrite("output_img/canny.jpg", edges)
else:
    print("Image not found.")
```

## 3. Thresholding

```python
# Apply simple binary thresholding to a grayscale image
if image is not None:
    greyscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    thresh_value = 127
    max_value = 255
    ret, dst = cv2.threshold(greyscale_image, thresh_value, max_value, cv2.THRESH_BINARY)
    cv2.imwrite("output_img/threshold.jpg", dst)
else:
    print("Image not found.")
```
