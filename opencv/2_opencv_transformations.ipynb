# OpenCV Image Transformations

This notebook covers resizing, cropping, rotation, and flipping techniques.

```python
import cv2
image = cv2.imread("input_img/sample.jpg")
```

## 1. Grayscale Conversion

```python
# Convert the image to grayscale and save
if image is not None:
    h, w, c = image.shape
    print(f"Image dimensions: {h}x{w}x{c}")

    grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    cv2.imwrite("output_img/grey.jpg", grey)
else:
    print("Image not found.")
```

## 2. Resizing

```python
# Resize the image to 300x300 pixels
if image is not None:
    resized = cv2.resize(image, (300, 300))
    cv2.imwrite("output_img/resized.jpg", resized)
else:
    print("Image not found.")
```

## 3. Cropping

```python
# Crop a specific Region of Interest (ROI)
if image is not None:
    # ROI coordinates: [y_start:y_end, x_start:x_end]
    x_start, y_start, x_end, y_end = 100, 100, 540, 327 
    
    cropped_img = image[y_start:y_end, x_start:x_end]
    cv2.imwrite("output_img/cropped.jpg", cropped_img)
else:
    print("Image not found.")
```

## 4. Rotation

```python
# Rotate the image 90 degrees clockwise
if image is not None:
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, -90, 1.0)
    rotated_image = cv2.warpAffine(image, M, (w, h))
    cv2.imwrite("output_img/rotated.jpg", rotated_image)
else:
    print("Image not found.")
```

## 5. Flipping

```python
# Apply different types of flips
if image is not None:
    flipped_v = cv2.flip(image, 0)
    cv2.imwrite("output_img/vertically_flipped.jpg", flipped_v)

    flipped_h = cv2.flip(image, 1)
    cv2.imwrite("output_img/horizontally_flipped.jpg", flipped_h)

    flipped_b = cv2.flip(image, -1)
    cv2.imwrite("output_img/both_flipped.jpg", flipped_b)
else:
    print("Image not found.")
```
