## 2. Basic Operations
Import OpenCV and load an image.

```python
import cv2
```

```python
# Read an image from the input directory
image = cv2.imread("input_img/sample.jpg")
```

```python
# Save the image to the output directory
if image is not None:
    success = cv2.imwrite("output_img/output.jpg", image)
    if success:
        print("Image saved successfully.")
    else:
        print("Failed to save the image.")
else:
    print("Image not found.")
```
