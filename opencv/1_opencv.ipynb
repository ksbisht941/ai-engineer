## 2. Basic Operations
Import OpenCV and load an image.

```python
import cv2
```

```python
# Read an image from the input directory
image = cv2.imread("input_img/sample.jpg")
```

```python
# Save the image to the output directory
if image is not None:
    success = cv2.imwrite("output_img/output.jpg", image)
    if success:
        print("Image saved successfully.")
    else:
        print("Failed to save the image.")
else:
    print("Image not found.")
```

## 3. Image Transformations
Grayscale conversion, resizing, cropping, and rotation.

```python
# Convert the image to grayscale and save
if image is not None:
    h, w, c = image.shape
    print(f"Image dimensions: {h}x{w}x{c}")

    grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    cv2.imwrite("output_img/grey.jpg", grey)
else:
    print("Image not found.")
```

```python
# Resize the image to 300x300 pixels
if image is not None:
    resized = cv2.resize(image, (300, 300))
    cv2.imwrite("output_img/resized.jpg", resized)
else:
    print("Image not found.")
```

```python
# Crop a specific Region of Interest (ROI)
if image is not None:
    # ROI coordinates: [y_start:y_end, x_start:x_end]
    x_start, y_start, x_end, y_end = 100, 100, 540, 327 
    
    cropped_img = image[y_start:y_end, x_start:x_end]
    cv2.imwrite("output_img/cropped.jpg", cropped_img)
else:
    print("Image not found.")
```

```python
# Rotate the image 90 degrees clockwise
if image is not None:
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    # Define rotation matrix and apply transformation
    M = cv2.getRotationMatrix2D(center, -90, 1.0)
    rotated_image = cv2.warpAffine(image, M, (w, h))
    cv2.imwrite("output_img/rotated.jpg", rotated_image)
else:
    print("Image not found.")
```

## 4. Image Flipping
Flipping images horizontally, vertically, or both.

```python
# Apply different types of flips
if image is not None:
    # flipCode: 0=Vertical, 1=Horizontal, -1=Both
    flipped_v = cv2.flip(image, 0)
    cv2.imwrite("output_img/vertically_flipped.jpg", flipped_v)

    flipped_h = cv2.flip(image, 1)
    cv2.imwrite("output_img/horizontally_flipped.jpg", flipped_h)

    flipped_b = cv2.flip(image, -1)
    cv2.imwrite("output_img/both_flipped.jpg", flipped_b)
else:
    print("Image not found.")
```

## 5. Image Filtering
Smoothing images with Gaussian blur.

```python
# Apply Gaussian blur with a (15, 15) kernel
if image is not None:
    blurred_img = cv2.GaussianBlur(image, (15, 15), 0)
    cv2.imwrite("output_img/blurred.jpg", blurred_img)
else:
    print("Image not found.")
```

## 6. Edge Detection
Detect edges with the Canny algorithm.

```python
# Detect edges using the Canny algorithm
if image is not None:
    edges = cv2.Canny(image, 100, 300)
    cv2.imwrite("output_img/canny.jpg", edges)
else:
    print("Image not found.")
```

## 7. Thresholding
Binarize images based on a threshold value.

```python
# Apply simple binary thresholding to a grayscale image
if image is not None:
    greyscale_image = cv2.imread("output_img/grey.jpg")
    
    thresh_value = 127
    max_value = 255

    # Pixels with value > 127 become 255, others become 0
    ret, dst = cv2.threshold(greyscale_image, thresh_value, max_value, cv2.THRESH_BINARY)
    cv2.imwrite("output_img/threshold.jpg", dst)
else:
    print("Image not found.")
```

## 8. Drawing Shapes
Drawing various shapes on images.

```python
# Draw a green line on the image
if image is not None:
    pt1 = (50, 300)
    pt2 = (300, 300)
    color = (0, 255, 0) # Green in BGR
    thickness = 4
    
    line_img = cv2.line(image.copy(), pt1, pt2, color, thickness)
    cv2.imwrite("output_img/line.jpg", line_img)
else:
    print("Image not found.")
```

```python
# Draw a green rectangle on the image
if image is not None:
    pt1 = (50, 50)
    pt2 = (250, 250)
    color = (0, 255, 0)
    thickness = 5
    
    rect_img = cv2.rectangle(image.copy(), pt1, pt2, color, thickness)
    cv2.imwrite("output_img/rectangle.jpg", rect_img)
else:
    print("Image not found.")
```

```python
# Draw a green circle on the image
if image is not None:
    center_coordinates = (256, 256)
    radius = 100
    color = (0, 255, 0)  # Green in BGR
    thickness = -1       # Filled circle
    
    circle_img = cv2.circle(image.copy(), center_coordinates, radius, color, thickness)

    cv2.imwrite("output_img/circle.jpg", circle_img)
else:
    print("Image not found.")
```

```python
# Draw a green circle on the image
if image is not None:
    # Define text parameters
    text_content = "Hello World!"
    position = (30, 250) # Bottom-left corner of text
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 1.5
    color = (255, 255, 255) # White in BGR
    thickness = 2
    line_type = cv2.LINE_AA # For smooth edges

    # Put the text on the image
    text_img = cv2.putText(image.copy(), text_content, position, font, font_scale, color, thickness, line_type)

    cv2.imwrite("output_img/text.jpg", text_img)
else:
    print("Image not found.")
```
